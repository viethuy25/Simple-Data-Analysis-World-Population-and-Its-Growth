```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
countries = pd.read_csv('countries.csv')
```

```python
us = countries[countries.country == "United States"]
china = countries[countries.country == "China"]
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(["United States","China"])
plt.ylabel('Population')
plt.xlabel('Year')
plt.show()
```

```python
population_1952 = countries[countries.year == 1952]
population_1957 = countries[countries.year == 1957]
population_1962 = countries[countries.year == 1962]
population_1967 = countries[countries.year == 1967]
population_1972 = countries[countries.year == 1972]
population_1977 = countries[countries.year == 1977]
population_1982 = countries[countries.year == 1982]
population_1987 = countries[countries.year == 1987]
population_1992 = countries[countries.year == 1992]
population_1997 = countries[countries.year == 1997]
population_2002 = countries[countries.year == 2002]
population_2007 = countries[countries.year == 2007]
```

```python
world_population_1952 = 0

for population in population_1952.iterrows():
    world_population_1952 += population[1].population
    
print (f'The world population is: {world_population_1952:1d}')
```

```python
year = [1952,1957,1962,1967,1972,1977,1982,1987,1992,1997,2002,2007]
world_population = []

x = 0
for y in year:
    count = 0
    for population in countries[countries.year == y].iterrows():
         count += population[1].population
    world_population.append(count)
    x += 1
```

```python
plt.plot(year, world_population)
plt.xlabel('Year')
plt.ylabel('World Population')
plt.show()
```

```python
growth_pop_each_country_2002_2007 = []
country_list = []

for population in population_2007.iterrows():
    country = population[1].country
    country_list.append(country)
    count = population[1].population
    
    for last_population in population_2002.iterrows():
        if last_population[1].country == country:
            count -= last_population[1].population
            break
    growth_pop_each_country_2002_2007.append(count/ 10**6)
```

```python
plt.plot(country_list,growth_pop_each_country_2002_2007)
plt.xlabel('Country (in order A-Z)')
plt.ylabel('Population growth (in million)')
plt.show()
```

```python
growth_world_pop_1952_2007 = [0]
year = [1952,1957,1962,1967,1972,1977,1982,1987,1992,1997,2002,2007]

pop_list = [population_1952,population_1957,population_1962,population_1967,population_1972,population_1977,population_1982,population_1987,population_1992,population_1997,population_2002,population_2007]

previous = next_ = None
l = len(pop_list)
for index, pop_year in enumerate(pop_list):
    
    if index >= 0 and index < (l-1):
        next_ = pop_list[index + 1]
        growth_pop_each_country = []
        count_total = 0
        sum_num = 0
        
        for population in pop_year.iterrows():
            country = population[1].country
            count_total = population[1].population
            
            for next_population in next_.iterrows():
                if next_population[1].country == country:
                    count_total -= last_population[1].population
                    break
            growth_pop_each_country.append(count_total/ 10**6)
            
        for num in growth_pop_each_country:
            sum_num += num
            
        growth_world_pop_1952_2007.append(sum_num)
            
    else:
        break
```

```python
plt.plot(year,growth_world_pop_1952_2007)
plt.xlabel('Year')
plt.ylabel('Population growth')
plt.show()
```

```python
average_growth_each_year = []

for x in growth_world_pop_1952_2007:
    average = x/142
    average_growth_each_year.append(average)
    
average_increase = 0
for increase in average_growth_each_year:
    average_increase += increase
    
average_increase= average_increase/12
fast_growth_country = []

for index, growth in enumerate(growth_pop_each_country_2002_2007):
    if (average_increase < growth):
        fast_growth_country.append(index)

country_name = []
i = 0
for index, country in population_2007.iterrows():
    if i in fast_growth_country:
        country_name.append(country.country)
        
    i+=1
        
country_name
```
